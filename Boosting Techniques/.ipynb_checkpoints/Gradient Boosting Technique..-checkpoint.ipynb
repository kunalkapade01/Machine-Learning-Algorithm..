# Gradient Boosting Implementation :

```python
# Import necessary library :
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Import sklearn library :
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

from sklearn.model_selection import GridSearchCV
```

```python
# Import warnings :
import warnings
warnings.filterwarnings("ignore")
```

- ùêÉùêöùê≠ùêö ùêãùê®ùêöùêù :

```python
# Loading the Data :
boston_house = pd.read_csv("D:\Dataset files_ML/boston_house_prices.csv")
boston_house
```

```python
# Spliting the data into x(Feature) & y(Target variable) :
x = boston_house.drop(["MEDV"], axis=1)
y = boston_house["MEDV"]
```

```python
x
```

```python
y
```

- ùêåùê®ùêùùêûùê• ùêìùê´ùêöùê¢ùêßùê¢ùêßùê† & ùêÑùêØùêöùê•ùêÆùêöùê≠ùê¢ùê®ùêß :

```python
# Initialize the model :
gbr = GradientBoostingRegressor(max_depth=2, n_estimators=10, learning_rate=1.0)
```

```python
# Split the data into train and test :
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=12)
```

```python
# Fit the model :
model = gbr.fit(X_train, y_train)
```

```python
# Trained the model on the test data :
y_pred = model.predict(X_test)
```

```python
# Check model r2 score :
r2_score(y_test, y_pred)
```

- ùêÖùêûùêöùê≠ùêÆùê´ùêû ùêàùê¶ùê©ùê®ùê´ùê≠ùêöùêßùêúùêû ùêèùê•ùê®ùê≠ :

```python
# Plot the important features :
feature_imp = model.feature_importances_

feature_imp = 100.0 * (feature_imp/feature_imp.max())
sorted_index = np.argsort(feature_imp)

# position on the features on bar plot.
position = np.arange(sorted_index.shape[0]) + 0.5

plt.barh(position, feature_imp[sorted_index], align="center")
plt.yticks(position, boston_house.columns[sorted_index])
plt.xlabel("Relative Importance")
plt.title("Freature Importance")
plt.show()
```

- ùêáùê≤ùê©ùêûùê´ùê©ùêöùê´ùêöùê¶ùêûùê≠ùêûùê´ ùêìùêÆùêßùê¢ùêßùê† :

```python
# Hyperparameter to tune their corresponding values :
parameter = {"learning_rate" : [0.15, 0.1, 0.10, 0.05],
            "n_estimators" : [100, 150, 200, 250],
            "max_depth" : [3, 4, 5],
            "min_samples_leaf" : [1, 2, 4],
            "min_samples_split" : [2, 5, 10]}

# Tuning the model as the estimator :
tuning = GridSearchCV(estimator=GradientBoostingRegressor(), param_grid=parameter, scoring="r2", cv=5)

# Fit the tuning model :
tuning.fit(X_train, y_train)

# Print the best hyperparameter and score found by GridSearchCV :
print("ùêÅùêûùê¨ùê≠ ùêáùê≤ùê©ùêûùê´ùê©ùêöùê´ùêöùê¶ùêûùê≠ùêûùê´: ", tuning.best_params_)
print("ùêÅùêûùê¨ùê≠ ùê´ùüê ùê¨ùêúùê®ùê´ùêû: ", tuning.best_score_)
```
