# Dicision_Tree Regression Model :

```python
# Import library :
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
# Import regression model :
from sklearn.datasets import make_regression
```

```python
# Create the data :
data = make_regression( n_samples=100, n_features=2, n_targets=1)
```

```python
data
```

```python
x = data[0]
y = data[1]
```

```python
# Plot the scatterplot :
plt.scatter(x[:,0], y, color='green', label="input")
plt.scatter(x[:,1], y, color='red', label="output")
plt.legend()
plt.show()
```

```python
# Create a Dataframe.
df = pd.DataFrame(x, columns=["col1","col2"])
df["target"] = y 
df.head()
```

```python
# checking the missing value in the dataframe.
df.isnull().sum()
```

```python
import sklearn.tree as tree
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
# Define regression model.
DTR = DecisionTreeRegressor(max_depth=3)
```

```python
# Fit the model :
DTR.fit(x, y)
```

```python
# Plot the tree :
plt.figure(figsize=(15,10))
tree.plot_tree(DTR)
plt.title("DecisionTreeRegression Model :")
plt.show()
```
