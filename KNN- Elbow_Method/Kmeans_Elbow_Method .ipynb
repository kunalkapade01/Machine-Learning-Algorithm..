```python
# Import Library :
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
# Import Load _Iris datasets :
from sklearn.datasets import load_iris
```

```python
# Load the data :
iris = load_iris()
```

```python
iris
```

```python
x = iris.data
```

```python
x
```

```python
y = iris.target
y
```

```python
# Plot the data :
%matplotlib inline
plt.scatter(x[y==0, 0], x[y==0, 1], color="green", label="Setosa")
plt.scatter(x[y==1, 0], x[y==1, 1], color="red", label="Versicolor")
plt.scatter(x[y==2, 0], x[y==2, 1], color="Black", label="Verginica")
plt.title("Sepal Length vs Sepal Width")
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.legend()
plt.show()
```

```python
%matplotlib inline
plt.scatter(x[y==0, 2], x[y==0, 3], color="green", label="Setosa")
plt.scatter(x[y==1, 2], x[y==1, 3], color="red", label="Versicolor")
plt.scatter(x[y==2, 2], x[y==2, 3], color="Black", label="Verginica")
plt.title("Sepal Length vs Sepal Width")
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.legend()
plt.show()
```

```python
%matplotlib inline
plt.scatter(x[y==0, 1], x[y==0, 2], color="green", label="Setosa")
plt.scatter(x[y==1, 1], x[y==1, 2], color="red", label="Versicolor")
plt.scatter(x[y==2, 1], x[y==2, 2], color="Black", label="Verginica")
plt.title("Sepal Length vs Sepal Width")
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.legend()
plt.show()
```

```python
from sklearn.cluster import KMeans
```

```python
# Define for n_clusters=3 :
km = KMeans(n_clusters=3)
```

```python
km
```

```python
# Fit the model & predict the clusters labels :
km.fit_predict(x)
```

```python
# Calculates within-clusters sum of squares :
km.inertia_
```

```python
# Find the error in the clusters :
error = []

for i in range(1, 11):
    k = KMeans(n_clusters=i)
    k.fit(x)
    error.append(k.inertia_)
```

```python
error
```

```python
# Plot the error :
plt.plot(range(1, 11), error, marker="*", markeredgecolor="black")
plt.show()
```
