```python
# Import library :
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 
import warnings
warnings.filterwarnings("ignore")
```

```python
from sklearn.datasets import make_classification
```

```python
# Make & create a data :
data = make_classification(n_samples=50, n_features=4, n_classes=2)
```

```python
data
```

```python
#Input :
x = data[0]
x = x[:,:2]
```

```python
#Output :
y = data[1]
y
```

```python
# New points with coordinates) :
new_p = np.array([0.54, -0.57])
```

```python
# Plot the points :
plt.scatter(x[y==0,0], x[y==0,1])
plt.scatter(x[y==1,0], x[y==1,1])
plt.scatter(new_p[0], new_p[1])
plt.show()
```

```python
# To calculate the Euclidean distance between the new pt and all other pts. :
dist = np.sqrt((new_p[0] - x[:,0])**2 + (new_p[1] - x[:,1])**2)
```

```python
from sklearn.datasets import make_moons
```

```python
# To generate a synthetic datasets : with 300 samples & Gaussian Noise.
moon = make_moons(n_samples=300, noise=0.15)
```

```python
moon 
```

```python
moon_df = pd.DataFrame(moon[0], columns=["col_1","col_2"])
moon_df["target"] = moon[1]
moon_df.head()
```

```python
# seperate the input and output :
x = moon[0]
y = moon[1]
```

```python
# plot the new data points :
plt.scatter(x[y==0,0], x[y==0,1])
plt.scatter(x[y==1,0], x[y==1,1])
plt.show()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_tr, x_te, y_tr, y_te = train_test_split(x, y, train_size=0.75)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
# Model create :
knn = KNeighborsClassifier(n_neighbors=int(np.sqrt(len(x_tr))))
```

```python
# Fit the model :
knn.fit(x_tr, y_tr)
```

```python
# score of the model :
knn.score(x_tr, y_tr)
```

```python
# Predict the model :
y_pred = knn.predict(x_te)
```

```python
y_pred
```

```python
# Evaluction the model :
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
Accuracy = accuracy_score(y_pred, y_te)
print(f"ùêÄùêúùêúùêÆùê´ùêöùêúùê≤: {Accuracy}")
```

```python
conf_matrix = confusion_matrix(y_pred, y_te)
print(f"ùêÇùê®ùêßùêüùêÆùê¨ùê¢ùê®ùêß ùêåùêöùê≠ùê´ùê¢ùê±: \n{conf_matrix}")    
```

```python
class_report = classification_report(y_pred, y_te)
print(f"ùêÇùê•ùêöùê¨ùê¨ùê¢ùêüùê¢ùêúùêöùê≠ùê¢ùê®ùêß ùêëùêûùê©ùê®ùê´ùê≠: {class_report}")
```
