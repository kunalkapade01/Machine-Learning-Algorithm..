```python
# Update the index :
df.reset_index(drop=True, inplace=True)
```

```python
df.info()
```

```python
#convert the data into standard scaler

from sklearn.preprocessing import StandardScaler
```

```python
std = StandardScaler()
```

```python
new_df = df.copy()
```

```python
new_df[["age","trestbps","chol","thalach","oldpeak"]] = std.fit_transform(new_df[["age","trestbps","chol","thalach","oldpeak"]])
```

```python
new_df
```

- 𝐀𝐩𝐩𝐥𝐲 𝐏𝐂𝐀 𝐌𝐨𝐝𝐞𝐥 :

```python
# Apply Principle component Analysis(PCA)

from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=3)
```

```python
x = pca.fit_transform(new_df.drop("target", axis=1))
```

```python
y = new_df["target"]
```

```python
x
```

```python
y
```

```python
pca.explained_variance_
```

```python
pca.get_covariance()
```

```python
var = df.iloc[34, :-1]
var
```

```python
pca.transform([var])
```

```python
from sklearn.svm import SVC
```

```python
svc = SVC(kernel="rbf")
```

```python
svc.fit(x, y)
```

```python
svc.score(x, y)
```

```python
new_pred = pca.transform([var])
```

```python
svc.predict(new_pred)
```
