# Resampling Method :

```python
# Import Necessary library :
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 

from sklearn.datasets import make_classification
from sklearn.utils import resample
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
# Make a data :
data = make_classification(n_samples=1000, n_classes=2,
                          weights=[0.8,0.2], random_state=12) 
```

```python
data
```

```python
# Create a Dataframe :
df = pd.DataFrame(data[0], columns=[f"feature_{i}"for i in range(len(data[0][0]))])
```

```python
df.head()
```

```python
# Add target column :
df["target"] = data[1]
```

```python
df.head()
```

```python
df["target"].value_counts()
```

```python
# Check majority :
majority = df[df["target"]==0]
majority.shape
```

```python
# Check minority :
minority = df[df["target"]==1]
minority.shape
```

```python
majority.drop("target", axis=1, inplace=True)
minority.drop("target", axis=1, inplace=True)
```

# Downsampling :

```python
Downsample = resample(majority, 
                     replace=False, n_samples=len(minority), random_state=12)
```

```python
down_df = pd.concat([minority, Downsample])
```

```python
down_df.shape
```

```python
down_df.head()
```

# Upsampling :

```python
upsample = resample(minority, 
                     replace=True, n_samples=len(majority), random_state=12)
```

```python
up_df = pd.concat([majority, upsample])
```
